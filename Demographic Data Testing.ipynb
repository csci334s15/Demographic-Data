```python
##### Testing the actual aggregation #####
# Empty Dictionary to hold all the zip codes and their respective average AGI value
avgAGIperZip = dict()
errorAGIStubs = []

# Loop over each zip code in the country
for zipcode in relevant_irs_data['zipcode'].unique():
    
    # Don't add 00000 and 99999 zip codes as they are error zipcodes and don't map to anywhere really
    if (zipcode == 0 or zipcode == 99999):
        continue
    else:
        # Temporary DataFrame with only entries for the current zipcode
        temp = relevant_irs_data[relevant_irs_data['zipcode'] == zipcode]
        # List of Number of Returns per AGI_STUB range
        counts = []
        # Total Number of Returns for the entire Zipcode
        totalCounts = sum(temp['N1'])
        # List of Avg AGI values per AGI_STUB range
        avg = []
        # List of the Avg AGI values after redistributing their respective number of returns
        # "undoing" the average calculation
        newAvg = []
        # Sum of the "undone" Avg AGI values
        zipAvg = 0

        # Add the Numbers of Returns to the array
        for eachCount in temp['N1']:
            counts.append(eachCount)

        # Add the Avg AGI value for each AGI_STUB
        for eachAvg in temp['A00100']:
            avg.append(eachAvg)

        # Add the zip code to the array if there are less than 6 stubs (error checking)
        if (len(avg) < 6):
            errorAGIStubs.append(zipcode)

        # "Undo" the average and add it to the new list
        for i in range(0, len(avg)):
            newAvg.append(avg[i]*counts[i])

        # Total the undone averages
        for each in newAvg:
            zipAvg += each

        # Complete the average with the new sum of all returns and the total number of returns
        finalZipAvg = zipAvg/totalCounts

        ### Add the actual average per zipcode into the dictionary
        # If the zip code is less than 5 digits, append a 0 to the front before you add it to the dictionary
        if (len(str(zipcode)) < 5):
            fixedZip = "0" + str(zipcode)
            avgAGIperZip[fixedZip] = round(finalZipAvg, 2)
        # If it's 5 digits, just add it
        else:
            avgAGIperZip[str(zipcode)] = round(finalZipAvg, 2)
```
